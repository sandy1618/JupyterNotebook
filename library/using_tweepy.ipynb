```python
# Authentication cell
consumer_key = MyAuth.consumer_key
consumer_secret = MyAuth.consumer_secret
access_token = MyAuth.access_token
access_token_secret = MyAuth.access_token_secret

auth = tw.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tw.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

# Define the search term and the date_since date as variables
search_words = "#Cybertruck" + "-filter:retweets"
# search_words = "#Cybertruck"
# date_since = "2018-11-16"

tweets = tw.Cursor(api.search,q = search_words,lang = 'en').items(5)
# tweets.text

# [print(tweet.created_at)for tweet in tweets]
users_locs = [[tweet.user.screen_name, tweet.user.location,tweet.geo,tweet.created_at,] for tweet in tweets]
users_locs
tweet_text = pd.DataFrame(data = users_locs, columns = ['user','location','geo','created_at'])
tweet_text
#     time.sleep(60)
```

### Structure
:

 **Tweet - Also referred to as a ‘Status’ object, has many ‘root-level’ attributes, parent of other objects.**

    User - Twitter Account level metadata. Will include any available account-level enrichments, such as Profile geo.

    Entities - Contains object arrays of #hashtags, @mentions, $symbols, URLs, and media.
    
    Extended Entities - Contains up to four native photos, or one video or animated GIF.
    
    Places - Parent to ‘coordinates’ object
    
Retweets
Retweets always contain two Tweet objects. The 'original' Tweet being Retweeted is provided in a "retweeted_status" object. 


also, a (new) location cannot be provided with a Retweet. While the 'original' Tweet may have geo-tagged, the Retweet "geo" and "place" objects will always be null.

geo-location of retweet is not available, but the user account creation location might be!
